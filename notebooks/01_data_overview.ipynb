## Interpretation (why this notebook matters)

This notebook answers two practical questions before you interpret any biology:

1. **What data exists for each figure in the paper?** (GEO sub-series map to Figs 1–5)
2. **Do we actually have the raw inputs locally to run downstream analysis notebooks?**

**How to read this page (talk-friendly):**

- **Bulk RNA-seq (GSE221693)** is the *fibroblast transcriptional response* arm (used to support the **MEOX1** activation story).
- **ChIP-seq (GSE221694)** is the *regulatory mechanism* arm at the **MEOX1** locus (H3K27ac = active regulatory chromatin; RELA = NF-κB binding).
- **CUT&RUN (GSE221695)** is the *BRD4 chromatin occupancy* arm at/near **Il1b** regulatory regions in myeloid cells.
- **scRNA/scATAC (GSE221698 / GSE221696)** are the *cell-type resolution* arms (Cell Ranger → Seurat/ArchR), usually requiring an x86 machine.

**Bottom line:** if the run folders/files aren’t present, downstream notebooks can’t compute — so this is your “inputs + scope check” for the replication.

```python
# Slide-friendly: downloaded runs by assay (ARM64 data/)
import matplotlib.pyplot as plt

assay_counts = (
    arm64_runs.assign(present=present)
    .groupby("assay")["present"].sum()
    .sort_values(ascending=True)
)

fig, ax = plt.subplots(figsize=(7, 3.5))
ax.barh(assay_counts.index, assay_counts.values)
ax.set_title("Local downloads under data/ (ARM64): runs by assay")
ax.set_xlabel("# run folders present")
for i, v in enumerate(assay_counts.values):
    ax.text(v + 0.1, i, str(int(v)), va="center")
plt.tight_layout()
plt.show()
```

## Key study findings

- **BRD4** in Cx3cr1+ myeloid cells regulates super-enhancers at *Il1b*
- **IL1B** signals to cardiac fibroblasts → induces *Meox1* and contractility
- **Anti-IL1B** treatment improves cardiac function and reduces fibroblast activation
